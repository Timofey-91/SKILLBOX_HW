## Задание №1

```python
x, y = symbols('x, y')
eq1 = -1 + 7 * y - 34
eq2 = 8 * x + 8 * y - 48
```

```python
linsolve([eq1, eq2], [x, y])
```

```python
numpy.linalg.solve(M1, V1)
```

```python
MSEx = diff(MSE, x)
MSEx
```

```python
MSEy = diff(MSE, y)
MSEy
```

```python
MSEz = diff(MSE, z)
MSEz
```
