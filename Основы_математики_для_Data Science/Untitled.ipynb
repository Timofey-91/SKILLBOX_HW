```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False, wrap_line=True, no_global=True)
```

```python
x = Symbol('x')
f = x**2 + 3*x - 4
f
```

```python
solve(f)
```

```python
solve_univariate_inequality(f>0, x)
```

```python
solve_univariate_inequality(f<0, x)
```

```python
plot(f)
```
