## Задание №2


```python
MSE = 0.25 * ((x2 * 1 + x1 * 2 - 5) ** 2 + (x2 * 2 + x1 * 4 - 10) ** 2 + (x2 * 3 + x1 * 7 - 8) ** 2 + (x2 * 5 + x1 * 3 - 6) ** 2)
MSE
```

```python
X1 = Matrix([[1, 2],
            [5, 3],
            [2, 4],
            [3, 7]])
X1
```

```python
Y1 = Matrix([5, 6, 10, 8])
Y1
```

```python
mul = (X1.T * X1) ** (-1) * X1.T * Y1
mul
```

```python
MSE.subs({x2:0, x1:0})
```

```python
MSE.subs({x2:mul[0], x1:mul[1]})
```
